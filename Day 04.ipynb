```python
aoc, lines, G, R, C = read_input(day, test=False)

# add_borders
G = ['_'*(R+2)] + [['_'] + g + ['_'] for g in G] + ['_'*(R+2)]
R = R+2
C = C+2

count = 0

def score(i,j):
    ss = 0
    for ii in range(i-1,i+2,1):
        for jj in range(j-1,j+2,1):
            if ii == i and jj == j:
                continue
            if G[ii][jj] == '@':
                ss += 1
    return ss

removed = None
while removed is None or len(removed) > 0:
    removed = []
    for i in range(1,C-1):
        for j in range(1,R-1):
            if G[i][j] == '@':
                s = score(i,j)
                if s < 4:
                    count +=1
                    removed.append((i,j))

    for i,j in removed:
        G[i][j] = '.'

print(count)
```
