```python
aoc, lines, G, R, C = read_input(day, test=False)

pairs = lines[0].split(',')

count = 0

for pair in pairs:
    ids = pair.split('-')
    for i in range(int(ids[0]), int(ids[1])+1):
        v = str(i)
        length = len(v)
        for l in range(1, length//2 + 1):
            splits = [v[n:n+l] for n in range(0, len(v), l)]
            if len(set(splits)) == 1:
                count += i
                break

print(count)

```
