```python
aoc, lines, G, R, C = read_input(day, test=False)

current = 50
counts = []

for i, l in enumerate(lines):

    prev = current
    if l[0] == 'R':
        current += int(l[1:])
    elif l[0] == 'L':
        current -= int(l[1:])


    revs = (prev // 100 - current // 100)


    if current % 100 == 0:
        if revs > 0:
            counts.append(revs + 1)
        elif revs < 0:
            counts.append(abs(revs))
        elif revs == 0:
            counts.append(1)
        else:
            assert False
    else:
        if revs > 0:
            counts.append(revs)
        elif revs < 0:
            counts.append(abs(revs))
        elif revs == 0:
            counts.append(0)
        else:
            assert False

    # Off by one if rotating left and starting at 0.
    if prev % 100 == 0 and l[0] == 'L':
        counts[-1] = counts[-1] - 1

    current = current % 100



print(sum(counts))
```
