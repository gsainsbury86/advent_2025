```python
aoc, lines, G, R, C = read_input(day, test=False)


SEARCH_LENGTH = 12

score = 0

def highest_with_at_least_left(input_string, L):
    # at least L left means only search up to -L.
    search_string = input_string[:-L] if L > 0 else input_string
    highest_first_digit = max([int(b) for b in search_string])
    index_of_highest_first_digit = input_string.index(str(highest_first_digit))
    return highest_first_digit, index_of_highest_first_digit


for bank in tqdm(lines):
    turned_on = ""
    current_index = 0
    for i in range(SEARCH_LENGTH-1,-1,-1):
        res, res_i = highest_with_at_least_left(bank[current_index:], i)
        current_index += res_i + 1
        turned_on+=str(res)
    score += int(turned_on)

print(score)
```
